# Load Model

### ERNIE

```python
import tensorflow as tf
import numpy as np
from transformers import AutoTokenizer, TFT5EncoderModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("michelecafagna26/t5-base-finetuned-sst2-sentiment")
t5_encoder = TFT5EncoderModel.from_pretrained('michelecafagna26/t5-base-finetuned-sst2-sentiment')
```

```python
t5_encoder.summary()
```

```python
# Freeze XLNet layers
for layer in t5_encoder.layers:
    layer.trainable = False
```

```python
# Define model inputs
input_ids = tf.keras.Input(shape=(256,), dtype=tf.int32, name='input_ids')
attention_mask = tf.keras.Input(shape=(256,), dtype=tf.int32, name='attention_mask')

# Pass inputs through the T5 encoder
encoder_outputs = t5_encoder(input_ids=input_ids, attention_mask=attention_mask)
sequence_output = encoder_outputs.last_hidden_state

# Pool the sequence output and add a classification layer
pooled_output = tf.keras.layers.GlobalAveragePooling1D()(sequence_output)
classifier = tf.keras.layers.Dense(1, activation='sigmoid', name='classifier')(pooled_output)

# Create the final model
t5_model = tf.keras.Model(inputs=[input_ids, attention_mask], outputs=classifier)

# Compile the model
t5_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
t5_model.summary()
```

```python
def preprocess_text(text, tokenizer, max_length=256):
    inputs = tokenizer(text, padding=True, truncation=True, return_tensors="tf", max_length=max_length)
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    return input_ids, attention_mask
```

```python
train_texts = train_abstract_df['text'].tolist()
val_texts = val_abstract_df['text'].tolist()

train_input_ids, train_attention_mask = preprocess_text(train_texts, tokenizer)
val_input_ids, val_attention_mask = preprocess_text(val_texts, tokenizer)

train_labels = np.array(train_abstract_df['label'].tolist())
val_labels = np.array(val_abstract_df['label'].tolist())

train_inputs = {
    "input_ids": train_input_ids,
    "attention_mask": train_attention_mask,
}

val_inputs = {
    "input_ids": val_input_ids,
    "attention_mask": val_attention_mask,
}
```

```python
train_inputs
```

```python
# Train the model
abstract_hist = t5_model.fit(
                    train_inputs, train_labels, 
                    validation_data=(val_inputs, val_labels), 
                    epochs=5,
                    batch_size=128
                )                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
```

```python
# Extract history data
history_dict = abstract_hist.history

# Create a DataFrame from the history dictionary
history_df = pd.DataFrame(history_dict)
```
