# Load Model

### XLNet

```python
import tensorflow as tf
import numpy as np

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
```

```python
import tensorflow as tf
from transformers import XLNetTokenizer, TFXLNetForSequenceClassification
```

```python
# Load XLNet tokenizer
tokenizer = XLNetTokenizer.from_pretrained('xlnet-base-cased')

# Load pre-trained XLNet model
xlnet_model = TFXLNetForSequenceClassification.from_pretrained('xlnet-base-cased')
```

```python
xlnet_model.summary()
```

```python
# Freeze XLNet layers
for layer in xlnet_model.layers:
    layer.trainable = False
```

```python
# Add classification layer on top of XLNet
classifier = tf.keras.layers.Dense(1, activation='sigmoid', name='classifier')

# Define model inputs
input_ids = tf.keras.Input(shape=(120,), dtype=tf.int32, name='input_ids')
attention_mask = tf.keras.Input(shape=(120,), dtype=tf.int32, name='attention_mask')

# Pass inputs through XLNet model
xlnet_output = xlnet_model({'input_ids': input_ids, 'attention_mask': attention_mask})

# Get logits output
logits = xlnet_output.logits

# Pass logits through classifier
output = classifier(logits)

# Create and compile the model
model = tf.keras.Model(inputs=[input_ids, attention_mask], outputs=output)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

```

```python
model.summary()
```

```python
# Tokenize and convert text data to input_ids and attention_mask
def preprocess_text(text):
    inputs = tokenizer(text, padding=True, truncation=True, return_tensors="tf", max_length=120)
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    return {"input_ids": input_ids, "attention_mask": attention_mask}
```

```python
# Preprocess training and validation data
train_inputs = preprocess_text(train_abstract_df['text'].tolist())
val_inputs = preprocess_text(val_abstract_df['text'].tolist())
```

```python
train_inputs
```

```python
model.fit(train_inputs, train_abstract_df['label'], validation_data=(val_inputs, val_abstract_df['label']), epochs=2, batch_size=32)
```

```python
test_inputs = preprocess_text(test_abstract_df['text'].tolist())
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_inputs, test_abstract_df['label'])
```

```python
# Make predictions
predictions = model.predict(test_inputs)
```

```python
predictions
```

```python
# Menghitung berapa banyak nilai probabilitas yang lebih besar dari 0.8
count_above_threshold = np.sum(predictions <= 0.5)

print("Jumlah nilai probabilitas di atas :", count_above_threshold)
```
